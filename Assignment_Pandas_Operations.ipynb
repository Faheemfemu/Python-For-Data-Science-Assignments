# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
auto_mobile.rename({'engine-location':'engine_location','num-of-doors':'num_of_doors'},inplace=True,axis=1)
auto_mobile.head()
```

```python
# Your code here. Use apply and lambda
def updated_price(engine_location , price):
  if engine_location == 'front':
    updated_price = price
  else:
    updated_price = price*2
  return updated_price
  
auto_mobile['updated_price'] = auto_mobile.apply(lambda x: updated_price(x['engine_location'], x['price']),axis=1)
auto_mobile.head()
```

```python
auto_mobile[['engine_location','price','updated_price']]
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers. 

```python
# Your code here. Use apply
auto_mobile[['num_of_doors']].value_counts()
#auto_mobile['num_of_doors'].replace({'four':'4','two':'2'},inplace=True )
#auto_mobile['num_of_doors']=auto_mobile['num_of_doors'].astype('int')
#auto_mobile.head()
def doors(num_of_doors):
  if num_of_doors=='four':
    doors= '4'
  elif num_of_doors=='two':    
     doors='2'
  else:
    doors=num_of_doors
  return doors
```

```python
auto_mobile['doors_value']=auto_mobile['num_of_doors'].apply(doors)
auto_mobile.head()
```
